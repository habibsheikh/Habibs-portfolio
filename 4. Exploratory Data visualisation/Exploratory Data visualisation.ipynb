```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
unrate = pd.read_csv('unrate.csv')
unrate.head()
```

```python
first_12 = unrate[0:12]
plt.plot(first_12['DATE'], first_12['VALUE'])
plt.show()
```

The above line chart x axis is messy cause items are overlapping. Lets fix this by rotating the x tick items by 90degrees!

```python
plt.plot(first_12['DATE'], first_12['VALUE'])
plt.xticks(rotation=90)
plt.show()      # now this is much bettter :) #
```

Now lets add axis labels to make the line chart more understandable.

```python
plt.plot(first_12['DATE'], first_12['VALUE'])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Rate, 1948')
plt.show()  # adding axis labels as well as a title has made the graph look clean!
```
